```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('preprocessed_data.csv',index_col=0)
df.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df.drop('status',axis=1),df['status'],
                                                test_size=0.2,random_state=34)
```

```python
xtrain.shape
```

```python
classifier = Sequential()
classifier.add(Dense(32, activation='relu', input_dim= 22))
classifier.add(Dense(16, activation='relu'))
classifier.add(Dense(1, activation='sigmoid'))
```

```python
classifier.summary()
```

```python
classifier.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
classifier.fit(xtrain, ytrain, epochs=400, verbose=1)
```

```python
ypred = np.round(classifier.predict(xtest)).astype(int)
```

```python
print(accuracy_score(ypred,ytest))
```

```python
# Using 400 epochs we were able to obtain an accuracy of about 95%
# The ANN achieved 100% accuracy at 120 epochs
```

```python
# Now we decrese the number of epochs to 120
classifier2 = Sequential()

classifier2.add(Dense(32, activation='relu', input_dim= 22))
classifier2.add(Dense(16, activation='relu'))
classifier2.add(Dense(1, activation='sigmoid'))

classifier2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

classifier2.fit(xtrain, ytrain, epochs=120, verbose=1)
```

```python
ypred2 = np.round(classifier2.predict(xtest)).astype(int)
```

```python
print(accuracy_score(ypred2,ytest))
```

```python
# The accuracy has increased to roughly 97%
# This is most likely due to the fact that the first classifier was 
# over-fitted to training data
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(ypred2,ytest))
```
