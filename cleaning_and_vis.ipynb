```python
# Let us now consider the individual distributions of the independent variables at the
# two levels of 'status' 
```

```python
independents = ['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']

for i in independents:
    plt.figure(figsize=(8,8))
    sns.boxplot(x = 'status', y = i, data=df)
    plt.show()
```

```python
# The box plots seem to corroborate the same notion as the correlation heat maps
# each of the independent variables taken on their own, seem to have an effect on the status
```
